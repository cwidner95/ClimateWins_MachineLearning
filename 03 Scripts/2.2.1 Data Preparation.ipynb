### 02. Data Wrangling

```python
# Dropping DATE and MONTH columns from weather dataset
df_weather = df_weather.drop(columns=['DATE','MONTH'])
df_weather.head()
```

```python
# Dropping DATE column from pleasant dataset
df_pleasant = df_pleasant.drop(columns=['DATE'])
df_pleasant.head()
```

```python
# Removing the Gdansk, Roma, and Tours stations as they do not have enough observations
removed_stations = ['GDANSK_cloud_cover', 'GDANSK_humidity', 'GDANSK_precipitation',
       'GDANSK_snow_depth', 'GDANSK_temp_mean', 'GDANSK_temp_min',
       'GDANSK_temp_max', 'ROMA_cloud_cover', 'ROMA_wind_speed',
       'ROMA_humidity', 'ROMA_pressure', 'ROMA_sunshine', 'ROMA_temp_mean',
       'TOURS_wind_speed', 'TOURS_humidity', 'TOURS_pressure',
       'TOURS_global_radiation', 'TOURS_precipitation', 'TOURS_temp_mean',
       'TOURS_temp_min', 'TOURS_temp_max']

df_weather = df_weather.drop(columns=removed_stations)
df_weather.shape
```

```python
# Dropping the snow_depth observations because they are missing multiple years
df_weather = df_weather.drop(df_weather.filter(regex='snow_depth').columns, axis=1)
df_weather.head()
```

```python
# Dropping the wind_speed observations because they are missing multiple years
df_weather = df_weather.drop(df_weather.filter(regex='wind_speed').columns, axis=1)
df_weather.head()
```

```python
df_weather.columns.tolist()
```

```python
# Copying cloud cover of Lujnljana to Kassel
pos = df_weather.columns.get_loc('KASSEL_humidity')
df_weather.insert(pos,'KASSEL_cloud_cover',df_weather['LJUBLJANA_cloud_cover'])
```

```python
# Copying pressure of Sonnblick to Munchen
pos = df_weather.columns.get_loc('MUNCHENB_global_radiation')
df_weather.insert(pos,'MUNCHENB_pressure',df_weather['SONNBLICK_pressure'])
```

```python
# Copying humidity of Oslo to Stockholm
pos = df_weather.columns.get_loc('STOCKHOLM_pressure')
df_weather.insert(pos,'STOCKHOLM_humidity',df_weather['OSLO_humidity'])
```
